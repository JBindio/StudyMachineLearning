```python
## 선형회귀(Linear Regression)
##### - 많이 사용되는 대표적인 회귀 알고리즘
##### - 비교적 간단하고, 성능이 뛰어나기 때문에 회귀모델 수행시 초반에 주로 진행합니다.
##### - 선형회귀의 역할 : 하나의 특성을 가장 잘 나타내는 직선을 찾아내는 것이 주 역할
##### - 직선을 "추세선" 이라고 합니다.
##### - 선형회귀 알고리즘 : 1차 방정식(y=ax+b)
##### - 선형회귀 클래스(모델) : LinearRegresin
##### - 사용되는 패키지 : sklearn.linear_model
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
perch_lenght = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )

perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 
     1000.0, 1000.0]
     )
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = \
    train_test_split(perch_lenght, perch_weight, random_state=42)
```

```python
train_input = train_input.reshape(-1,1)
test_input = test_input.reshape(-1,1)
```

```python
lr.fit(train_input, train_target)
```

```python
lr.score(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
nfish = [[50]]
```

```python
lr.predict(nfish)
```

```python
# 일차방정식의 기울기와 y절편값 확인하기
# 일차방정식 : y = ax + b
# a : 기울기(coef_)
# b : y절편(intercept_)

print(lr.coef_, lr.intercept_)
```

```python
# 추세선을 이용한 예측데이터 예측 가능여부 확인하기
    # - 예측데이터로 사용할 수 있는지 확인
    # - 산점도를 이용한 확인
    # - 훈련데이터, 예측데이터 산점도로 그리기
    
# 선형 회귀 모델 산점도
fig = plt.figure(figsize=(20,10))
plt.rc('font', family = 'Malgun Gothic', size = 18)
plt.title('선형회귀모델 산점도')

plt.scatter(train_input, train_target, marker='D',s=30, c='navy')
plt.scatter(nfish, lr.predict(nfish), marker='*', s=300, c='red')

# 추세선 그리기
# 추세선 : 선형회귀 알고리즘(1차방정식)이 훈련데이터에서 찾은 최적의 직선
plt.plot([15,50],[lr.coef_ * 15 + lr.intercept_,
                  lr.coef_ * 50 + lr.intercept_], color = 'darkorange')

plt.rcParams['lines.linewidth'] = 2

plt.xlabel('농어 길이(cm)')
plt.ylabel('농어 무게(kg)')
plt.show()
```

```python
# 과적합 여부 확인하기
print(lr.score(train_input,train_target))
print(lr.score(test_input, test_target))
```

```python
# 추세선을 이용한 예측데이터 예측 가능여부 확인하기
    # - 예측데이터로 사용할 수 있는지 확인
    # - 산점도를 이용한 확인
    # - 훈련데이터, 예측데이터 산점도로 그리기
    
# 선형 회귀 모델 산점도
fig = plt.figure(figsize=(20,10))
plt.rc('font', family = 'Malgun Gothic', size = 18)
plt.title('선형회귀모델 산점도')

plt.scatter(train_input, train_target, marker='D',s=30, c='navy')
plt.scatter(nfish, lr.predict(nfish), marker='*', s=300, c='red')

# 추세선 그리기
# 추세선 : 선형회귀 알고리즘(1차방정식)이 훈련데이터에서 찾은 최적의 직선
plt.plot([0,50],[lr.coef_ * 0 + lr.intercept_,
                  lr.coef_ * 50 + lr.intercept_], color = 'darkorange')

plt.rcParams['lines.linewidth'] = 2

plt.xlabel('농어 길이(cm)')
plt.ylabel('농어 무게(kg)')
plt.show()
```

```python
# <해석>
# - 테스트 결과가 다소 낮게 나왔으며, 훈련데이터는 높은편
# - 다소 차이는 있으나 우수한 모델로 평가 할 수 있음
# - 단, 훈련데이터와 테스트데이터의 결정계수 차이는 과대적합을 의심해 봐야 함

# 그래프 왼쪽 하단의 산점도 분포가 특이함
    # - 약간의 ronud(곡선)형태
    # - 추세선을 왼쪽 하단으로 직선을 연장했을때 y축이 0이하로 되며 무게는 0보다 작을 수 없음
    # - 음수를 예측하기 때문에 잘못된 모델.
```

```python
train_poly = np.column_stack((train_input**2, train_input))
train_poly.shape
```

```python
test_poly = np.column_stack((test_input**2, test_input))
test_poly.shape
```

```python
# - 객체(모델) 생성
# - 훈련모델 생성
# - 예측데이터 : 길이 50으로 무게 예측
# - 계수 확인 (coef_, intercept_)
# - 산점도 그리기 : 훈련데이터, 예측데이터, 추세선
```

```python
lr_poly = LinearRegression()
```

```python
lr_poly.fit(train_poly, train_target)
```

```python
lr_poly.score(train_poly, train_target)
```

```python
lr_poly.score(test_poly, test_target)
```

```python
nfish_poly = [[50**2, 50]]
```

```python
lr_poly.predict(nfish_poly)
```

```python
print(lr_poly.coef_, lr_poly.intercept_)
```

```python
lr_poly.coef_[0]
```

```python
## 추세선을 이용한 예측데이터 예측 가능여부 확인하기
    # - 예측데이터로 사용할 수 있는지 확인
    # - 산점도를 이용한 확인
    # - 훈련데이터, 예측데이터 산점도로 그리기
    
# 다항 회귀 모델 산점도
fig = plt.figure(figsize=(20,10))
plt.rc('font', family = 'Malgun Gothic', size = 18)
plt.title('다항 회귀모델 산점도')

plt.scatter(train_input, train_target, marker='D',s=30, c='navy')
plt.scatter(nfish_poly[0][1], lr_poly.predict(nfish_poly), marker='*', s=300, c='red')

# # 추세선 그리기
# # 추세선 : 다항회귀 알고리즘(2차방정식)이 훈련데이터에서 찾은 최적의 곡선
point = np.arange(15,51)
plt.plot(point, lr_poly.coef_[0] * point **2 + lr_poly.coef_[1] * point + lr_poly.intercept_ , color = 'darkorange',zorder =0)

plt.rcParams['lines.linewidth'] = 2

plt.xlabel('농어 길이(cm)')
plt.ylabel('농어 무게(kg)')
plt.show()
```

```python
# 과적합 확인하기
print('  훈련 :',lr_poly.score(train_poly, train_target))
print('테스트 :',lr_poly.score(test_poly, test_target))

# 테스트 데이터가 결정계수가 조금 더 높음
# 현업에서는 사용가능하지만
# 학문적으로는 인정하지 않는다 (논문, 연구, 특허)
```

```python
# <해석>
# 전체적으로 정확도는 조금 낮아졌습니다.
# 그러나, 훈련/테스트의 결정계수는 유사하게 가까워 졌기에,
# 우수한 모델이라고 할 수 있습니다.

# 다만 아직 훈련이 테스트보다 낮은 과소적합을 보입니다.
# <과소적함 해결방안>
    # - 복잡도를 줄 수 있는 모델 찾기 > 다중회귀분석
```
